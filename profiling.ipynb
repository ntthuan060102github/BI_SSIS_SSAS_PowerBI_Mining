```python
import pandas as pd
import dateutil.parser
from datetime import datetime
```

```python
df = pd.read_excel("./data_source/supermarket_sales.xlsx")
df
```

```python
# Product ID
product_id_counter = df.groupby(["Branch", "Customer type", "ProductID", "Date", "Time", "Payment"]).size().reset_index(name='Count')
product_id_counter[product_id_counter['Count'] == 1]
```

```python
# Invoice ID
invoice_id_counter = df.groupby(["Invoice ID"]).size().reset_index(name='Count')
invoice_id_counter[invoice_id_counter['Count'] > 1]
```

```python
# Branch
branchs = df["Branch"].drop_duplicates()
branchs
```

```python
# Customer Type
customer_types = df["Customer type"].drop_duplicates()
customer_types
```

```python
# Gender
genders = df["Gender"].drop_duplicates()
genders
```

```python
# Test
purchase_counter = df.groupby(["ProductID"]).size().reset_index(name='Count')
```

```python
# Quantity
invalid_quantity_counter = df[(df["Quantity"] < 1) | (~df["Quantity"].astype(int).eq(df["Quantity"]))]
invalid_quantity_counter
```

```python
# Tax 5%
invalid_tax_counter = df[df["Tax 5%"] < 0]
invalid_tax_counter
```

```python
# Total
invalid_total_counter = df[df["Total"] < 0]
invalid_total_counter
```

```python
# Date
purchage_date_out_of_range = df[~(df['Date'].dt.year == 2019) & (df['Date'].dt.month.between(1, 3))]
purchage_date_out_of_range
assert len(purchage_date_out_of_range) == 0, f"Has {len(purchage_date_out_of_range)} invalid rows"
```

```python
# Time
df["Purchage Date Time"] = datetime.combine(df['Date'][0].date(), df['Time'][0])
purchage_time_out_of_range = df[~df["Purchage Date Time"].dt.hour.between(10, 21)]
assert len(purchage_time_out_of_range) == 0, f"Have {len(purchage_time_out_of_range)} invalid rows"
```

```python
# Payment
payment_types = df["Payment"].drop_duplicates()
payment_types
```

```python
# cogs, gross margin percentage, gross income
invalid_cost_verifier = df[
    (df["cogs"] < 0)
    | (df["gross margin percentage"] < 0)
    | (df["gross income"] < 0)
]
invalid_cost_verifier
```

```python
# Rating
rating_out_of_range = df[(df["Rating"] > 10.) | (df["Rating"] < 1.)]
rating_out_of_range
```
